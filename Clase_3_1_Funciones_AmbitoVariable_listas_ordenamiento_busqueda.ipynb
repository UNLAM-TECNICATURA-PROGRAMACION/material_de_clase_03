## üîß 1. Funciones en Python

Una **funci√≥n** es un bloque de c√≥digo que se puede reutilizar y ejecutar con diferentes datos.

```python
def saludar(nombre):
    print(f"Hola, {nombre}!")  # Imprime un saludo personalizado

saludar("Alex")
saludar("Zelda")
```

## üß† 2. √Åmbito de Variables

El **√°mbito** define d√≥nde se puede acceder a una variable: **local** (dentro de una funci√≥n) o **global** (fuera).

```python
x = 10  # variable global

def modificar():
    x = 5  # variable local
    print("Dentro de la funci√≥n:", x)

modificar()
print("Fuera de la funci√≥n:", x)
```

## üìö 3. Listas (Arrays)

Las **listas** almacenan m√∫ltiples valores y permiten muchas operaciones √∫tiles.

```python
personajes = ["Mario", "Link", "Samus"]
print("Segundo personaje:", personajes[1])

personajes.append("Kirby")  # Agregar al final
print("Lista actualizada:", personajes)
```

## üîÉ 4. Ordenamiento

### ‚úÖ Ordenar con `sort()`

```python
niveles = [3, 1, 4, 2]
niveles.sort()
print("Ordenado:", niveles)
```

### ‚úÖ Ordenar sin modificar la lista original con `sorted()`

```python
original = [8, 3, 5]
ordenado = sorted(original)
print("Original:", original)
print("Ordenado:", ordenado)
```

## üîç 5. B√∫squeda

### B√∫squeda Secuencial

```python
def buscar(lista, objetivo):
    for i, valor in enumerate(lista):
        if valor == objetivo:
            return i  # Encontrado
    return -1  # No encontrado

print(buscar([10, 20, 30], 20))  # ‚Üí 1
print(buscar([10, 20, 30], 25))  # ‚Üí -1
```

## üöÄ Desaf√≠o Final Resuelto

Crea una funci√≥n que:
- Reciba una lista de puntajes
- Ordene de mayor a menor
- Busque un puntaje espec√≠fico y devuelva su posici√≥n

```python
def procesar_puntajes(puntajes, objetivo):
    ordenados = sorted(puntajes, reverse=True)
    print("Puntajes ordenados:", ordenados)

    if objetivo in ordenados:
        indice = ordenados.index(objetivo)
        print(f"El puntaje {objetivo} est√° en la posici√≥n {indice} del ranking.")
    else:
        print(f"El puntaje {objetivo} no est√° en el ranking.")

procesar_puntajes([320, 540, 200, 450], 450)
```
